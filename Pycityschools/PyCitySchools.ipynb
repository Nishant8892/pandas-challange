```python
 #Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

school_data_complete.head()
```

```python
#create bins and labels and cut per school regarding spending ranges

budget_bins = [0,585,630,645,680]
budget_labels = ["<$585", "$585-630","$630-645","$645-680"]

spending_summary_df = school_summary_df.copy()

spending_summary_df["Spending Ranges (Per Student)"] = pd.cut(x= per_student_budget, bins = budget_bins, 
                                            labels = budget_labels)
spending_summary_df 
```

```python
# group DF to average maths/reading score and % passing score for maths/reading/overall

Average_maths_spending = spending_summary_df.groupby(["Spending Ranges (Per Student)"]).mean()["Average Maths Score"]
Average_reading_spending = spending_summary_df.groupby(["Spending Ranges (Per Student)"]).mean()["Average Reading Score"]
percentage_maths_spending = spending_summary_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Maths"]
percentage_reading_spending= spending_summary_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Reading"]
percentage_overall_spending= spending_summary_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Overall Passing"]
```

```python
# create DF to summarise spending per student

final_spending_summary = pd.DataFrame({"Average Maths Score": round(Average_maths_spending,2),
                              "Average Reading Score": round(Average_reading_spending,2),
                              "% Passing Maths": round(percentage_maths_spending,2),
                            "% Passing Reading": round(percentage_reading_spending,2),
                            "% Overall Passing": round(percentage_overall_spending,2)         })
 

final_spending_summary
```

```python
#complete similar action as above but now we are comparing school size to \
#average maths/reading score and % passing score for maths/reading/overall

student_bins = [0,1000,2000,5000]
student_labels = ["Small (<1000)", "Medium (1000-2000)","Large (2000-5000)"]

school_size_df = school_summary_df.copy()

school_size_df["School Size"] = pd.cut(x= total_student_count, bins = student_bins, 
                                            labels = student_labels)

Average_maths_size = school_size_df.groupby(["School Size"]).mean()["Average Maths Score"]
Average_reading_size = school_size_df.groupby(["School Size"]).mean()["Average Reading Score"]
percentage_maths_size = school_size_df.groupby(["School Size"]).mean()["% Passing Maths"]
percentage_reading_size= school_size_df.groupby(["School Size"]).mean()["% Passing Reading"]
percentage_overall_size= school_size_df.groupby(["School Size"]).mean()["% Overall Passing"] 

final_school_size_summary = pd.DataFrame({"Average Maths Score": round(Average_maths_size,2),
                              "Average Reading Score": round(Average_reading_size,2),
                              "% Passing Maths": round(percentage_maths_size,2),
                            "% Passing Reading": round(percentage_reading_size,2),
                            "% Overall Passing": round(percentage_overall_size,2)         })
 

final_school_size_summary
```
